```python
from PIL import Image
import os
import json
```

```python
SOURCE_DIR = "raw_data/simple_floor"
TARGET_DIR = "simple_floor_coco"
VAL_PCT = 0.2
```

```python
categories = []
os.listdir(SOURCE_DIR)
with open(SOURCE_DIR + '/annotation_definitions.json') as f:
    a = json.load(f)
#print(a)
for i in a['annotationDefinitions'][0]['spec']:
    #print(i)
    #print(i)
    categories.append({
        'id': i['label_id'], 
        'name': i['label_name'], 
        'supercategory': 'all'
    })
categories
```

```python
images = []
annotations = []
img_id = 0
ann_id = 0

for i in os.listdir(SOURCE_DIR + '/sequence.0'):
    #print(i)
    if i.endswith('json') and not i.startswith('annotations'):
        #print(i)
        with open(SOURCE_DIR + '/sequence.0/' + i, 'r') as f:
            d = json.load(f)
            #for j in d['captures']:
            #    annotations.append(d['captures'][j]['annotations'])
            #print(d.keys())
            #print(d['captures'][0].keys())
            #print(d['captures'][0]['annotations'])
            for j in d['captures'][0]['annotations'][0].get('values', []):
                #print(j)
                annotations.append({
                    'id': ann_id,
                    'image_id': img_id, 
                    'bbox': j['origin'] + j['dimension'], 
                    'category_id': j['labelId'], 
                    'iscrowd': 0,
                    'area': j['dimension'][0] * j['dimension'][1],
                    'segmentation': [j['origin'][0],
                                     j['origin'][1],
                                     j['origin'][0],
                                     j['origin'][1] + j['dimension'][1],
                                     j['origin'][0] + j['dimension'][0],
                                     j['origin'][1] + j['dimension'][1],
                                     j['origin'][0] + j['dimension'][0],
                                     j['origin'][1]
                                    ]
                })
                ann_id += 1
            #print(d['captures'][0]['annotations'][0]['values'][0])
            #print(len(d['captures']))
            #break
        img = Image.open(SOURCE_DIR + '/sequence.0/' + d['captures'][0]['filename'])
        images.append({
            'id': img_id, 
            'file_name': d['captures'][0]['filename'],
            'width': img.width,
            'height': img.height,
        })
        img.close()
        img_id += 1
```

```python
counts = {i['id'] : 0 for i in images}
for i in annotations:
    counts[i['image_id']] += 1
counts
```

```python
images = [i for i in images if counts[i['id']] > 1]
len(images)
```

```python
import random

random.shuffle(images)

split_index = int(VAL_PCT * len(images))
val_images = images[:split_index]
train_images = images[split_index:]

val_annotations = []
train_annotations = []
for i in annotations:
    if i['image_id'] in [j['id'] for j in train_images]:
        train_annotations.append(i)
    elif i['image_id'] in [j['id'] for j in val_images]:
        val_annotations.append(i)
len(val_annotations), len(train_annotations)
```

```python
val_out = {
    'annotations': val_annotations,
    'images': val_images,
    'info': {},
    'license': {},
    'categories': categories
}

train_out = {
    'annotations': train_annotations,
    'images': train_images,
    'info': {},
    'license': {},
    'categories': categories
}
```

```python
import shutil
```

```python
os.makedirs(TARGET_DIR + '/train2017')
os.makedirs(TARGET_DIR + '/val2017')
```

```python
for i in os.listdir(SOURCE_DIR + '/sequence.0'):
    if i.endswith('png'):
        if i in [j['file_name'] for j in train_images]:
            shutil.copyfile(SOURCE_DIR + '/sequence.0/' + i, TARGET_DIR + '/train2017/' + i)
        elif i in [j['file_name'] for j in val_images]:
            shutil.copyfile(SOURCE_DIR + '/sequence.0/' + i, TARGET_DIR + '/val2017/' + i)
```

```python
with open(TARGET_DIR + '/annotations_train.json', 'w') as f:
    json.dump(train_out, f)
with open(TARGET_DIR + '/annotations_val.json', 'w') as f:
    json.dump(val_out, f)
```
